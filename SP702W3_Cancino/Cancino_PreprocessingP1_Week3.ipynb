# Data Preprocessing Part 1

### Give an initial characterization of entire dataframe.

```python
import pandas as pd
```

```python
# Importing the sales.csv in the notebook
sales = pd.read_csv('Cancino_sales.csv')
#Giving an initial characterization of entire dataframe.
sales.describe(include='all')
```

### Remove duplicate values

```python
# Removing the duplicate values in dataframe.
sales = sales.drop_duplicates()
sales
```

### Express Price in thousands

```python
sales.Price = (sales.Price)/1000
```

```python
sales.rename(columns = {'Price':'Price(K)'} , inplace=True)
```

```python
sales
```

### Select name, city, states and country data of Product1 by customers from united states and paid thru Amex or Diners. Store selected data in empty.csv file

```python
data = sales[['Name', 'City' , 'State' , 'Country']]
```

```python
data_new = data[(sales['Payment_Type'] == 'Amex') | (sales['Payment_Type'] == 'Diners') & (sales['Product'] == 'Product1') & (sales['Country']=='United States')] 
data_new.reset_index(inplace=True , drop=True)
data_new.to_csv('Cancino_empty.csv' , index=False)
data_new
```
